## Step 5: Descriptive Statistics
Calculate and interpret measures of central tendency and dispersion.
Summarize key statistics for numerical variables.

```python
# Descriptive statistics
print(df.describe())

# Central tendency and dispersion for math scores
mean_Age= df['Age'].mean()
median_Age = df['Age'].median()
std_Age = df['Age'].std()

print(f"Mean Age: {mean_Age}")
print(f"Median Age: {median_Age}")
print(f"Standard Deviation of Age: {std_Age}")
```

## Step 6: Identify Patterns and Relationships
Use correlation analysis to identify relationships between variables.
Perform bivariate and multivariate analyses.

```python
# Correlation between math score and reading score
correlation_Age_PurchaseAmount = df['Age'].corr(df['PurchaseAmount'])
print(f"Correlation between Age and Purchase Amount: {correlation_Age_PurchaseAmount}")

# Grouped analysis of average scores by parental level of education and gender
grouped_scores = df.groupby(['Size', 'Gender'])[['Age', 'Review Rating', 'Previous Purchases']].mean().reset_index()
print(grouped_scores)
```

## Step 7: Outlier Detection
Identify and handle outliers using visualizations and statistical methods.

```python
# Box plot for outlier detection in math scores
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Age'])
plt.title('Box Plot for Age')
plt.show()

# Handling outliers (example: using IQR method)
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1

# Define outliers
outliers = df[(df['Age'] < (q1 - 1.5 * iqr)) | (df['Age'] > (q3 + 1.5 * iqr))]
print(outliers)
```

## Step 8: Hypothesis Testing
Formulate hypotheses based on initial observations and test them using statistical methods.

```python
from scipy.stats import ttest_ind

# Hypothesis: Male and female students have different math scores
Male_Age = df[df['Gender'] == 0]['Age']
Female_Age = df[df['Gender'] == 1]['Age']

t_stat, p_value = ttest_ind(Male_Age, Female_Age)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
